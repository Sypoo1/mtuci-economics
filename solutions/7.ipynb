```python

# Данные задачи
SP_ost = 600.0      # Условно-постоянные расходы, тыс. руб.
C_ed = 0.5          # Цена за единицу продукции, тыс. руб.
C_ed_per = 0.3      # Условно-переменные расходы на единицу, тыс. руб.

# Вычисление критического объёма (Qкр) в натуральном выражении
Q_kr = SP_ost / (C_ed - C_ed_per)


print("Задача 7.1: Критический объём производства")
print(f"  - Qкр (ед. продукции) = {Q_kr:.2f}")


```

```python

# Данные задачи
MZ = 0.1            # Материальные затраты на единицу, тыс. руб.
SP_ed = 0.35        # Полная себестоимость единицы продукции, тыс. руб.
annual_quantity = 6000
PFSR = 2500.0       # Среднегодовая стоимость производственных фондов, тыс. руб.
R_pr = 0.15         # Рентабельность предприятия
NDS_rate = 0.20     # НДС = 20%

# 1) Находим общую (годовую) себестоимость
SP_total = SP_ed * annual_quantity  # тыс. руб.

# 2) Рассчитываем уровень рентабельности по себестоимости для одного вида продукции (RП)
R_P = (PFSR * R_pr) / SP_total

# 3) Оптовая цена предприятия (без НДС)
C_opt_predpr = SP_ed * (1 + R_P)

# 4) Отпускная цена предприятия (с НДС) по формуле (7.5)
C_otp_predpr = C_opt_predpr + (C_opt_predpr - MZ) * NDS_rate

print("Задача 7.2: Отпускная цена предприятия")
print(f"  - Полная себестоимость годового объёма (SP_total), тыс. руб.: {SP_total:.2f}")
print(f"  - Уровень рентабельности по себестоимости (RП): {R_P:.4f}")
print(f"  - Оптовая цена (без НДС), тыс. руб.: {C_opt_predpr:.4f}")
print(f"  - Отпускная цена (с НДС), тыс. руб.: {C_otp_predpr:.4f}\n")
```
