```python

# Исходные данные
q = 10000            # Годовой выпуск продукции, ед.
S = 80.0            # Себестоимость единицы, руб.
markup = 0.25       # Наценка 25%
CO = 50000.0        # Среднегодовая стоимость оборотных средств, руб.
T_p = 360           # Длительность периода в днях

# 1) Цена одного изделия
price = S * (1 + markup)

# 2) Годовой объём реализации
R_p = q * price

# 3) Коэффициент оборачиваемости
K_o = R_p / CO

# 4) Средняя продолжительность одного оборота
T_o = T_p / K_o

# 5) Коэффициент загрузки (закрепления)
K_z = CO / R_p

# Вывод результатов
print("Задача 8.1: Показатели оборачиваемости оборотных средств\n")
print(f"  - Цена одного изделия, руб.: {price:.2f}")
print(f"  - Годовой объём реализации (R_p), руб.: {R_p:.2f}")
print(f"  - Коэффициент оборачиваемости (K_o): {K_o:.2f}")
print(f"  - Средняя продолжительность одного оборота (T_o), дней: {T_o:.2f}")
print(f"  - Коэффициент загрузки (K_z): {K_z:.4f}\n")


```

```python

# Исходные данные
CO = 100000.0            # Сумма нормированных оборотных средств, руб.
T_o_otchet = 35.0        # Длительность одного оборота в отчетном году, дней
increase_rate = 0.05     # Рост объема реализации (5%)
T_p = 360.0              # Период (дней в году)

# 1) Коэффициент оборачиваемости в отчетном году
K_o_otchet = T_p / T_o_otchet

# 2) Объем реализованной продукции в отчетном году
R_p_otchet = K_o_otchet * CO

# 3) Объем реализации в будущем году
R_p_bud = R_p_otchet * (1 + increase_rate)

# 4) Коэффициент оборачиваемости в будущем году
K_o_bud = R_p_bud / CO

# 5) Новая длительность одного оборота
T_o_bud = T_p / K_o_bud

# 6) На сколько дней сократится время оборота
delta_days = T_o_otchet - T_o_bud

# Вывод результатов
print("Задача 8.2: Сокращение длительности одного оборота в будущем году\n")
print(f"  - Коэффициент оборачиваемости (отчетный) K_o_otchet: {K_o_otchet:.2f}")
print(f"  - Объём реализации (отчетный) R_p_otchet, руб.: {R_p_otchet:.2f}")
print(f"  - Объём реализации (будущий) R_p_bud, руб.: {R_p_bud:.2f}")
print(f"  - Коэффициент оборачиваемости (будущий) K_o_bud: {K_o_bud:.2f}")
print(f"  - Новая длительность одного оборота T_o_bud, дней: {T_o_bud:.2f}")
print(f"  - Сокращение времени оборота (дней): {delta_days:.2f}\n")

```
